```python
# Ignore warnings
import warnings
warnings.filterwarnings('ignore')

# Import Path from pathlib
from pathlib import Path

# Import yfinance package
import yfinance as yf

# Import pandas package
import pandas as pd

# Import pretty printing
from pprint import pprint

# Load the data issuers data
issuers = pd.read_csv("stock_technical_indicators_last_5_years.csv")

# View the issuers data -> Issuer Name and Security Id
issuers.head()

# Pull the data for the first security
ticker = issuers.iloc[0]['ticker']
stock_data = yf.Ticker(ticker)

# Extract full of the stock
stock_data_info = stock_data.info

# Extract only the important information
stock_data_info = {
    'Basic Information': {
        'symbol': stock_data_info.get('symbol', 'N/A'),
        'longName': stock_data_info.get('longName', 'N/A'),
        'currency': stock_data_info.get('currency', 'N/A'),
        'exchange': stock_data_info.get('exchange', 'N/A')
    },
    'Market Data': {
        'currentPrice': stock_data_info.get('currentPrice', 'N/A'),
        'previousClose': stock_data_info.get('previousClose', 'N/A'),
        'open': stock_data_info.get('open', 'N/A'),
        'dayLow': stock_data_info.get('dayLow', 'N/A'),
        'dayHigh': stock_data_info.get('dayHigh', 'N/A'),
        'regularMarketPreviousClose': stock_data_info.get('regularMarketPreviousClose', 'N/A'),
        'regularMarketOpen': stock_data_info.get('regularMarketOpen', 'N/A'),
        'regularMarketDayLow': stock_data_info.get('regularMarketDayLow', 'N/A'),
        'regularMarketDayHigh': stock_data_info.get('regularMarketDayHigh', 'N/A'),
        'fiftyTwoWeekLow': stock_data_info.get('fiftyTwoWeekLow', 'N/A'),
        'fiftyTwoWeekHigh': stock_data_info.get('fiftyTwoWeekHigh', 'N/A'),
        'fiftyDayAverage': stock_data_info.get('fiftyDayAverage', 'N/A'),
        'twoHundredDayAverage': stock_data_info.get('twoHundredDayAverage', 'N/A')
    }
}

# View the stock info
pprint(stock_data_info)

# Extract the data for last 2 years with daily interval
stock_data_hist = stock_data.history(period="2y", interval="1d")

# Clean the data to keep only the 'Close' column
stock_data_hist = stock_data_hist[['Close']]

# Change frequency to day and fill missing values
stock_data_hist = stock_data_hist.asfreq('D', method='ffill').fillna(method='ffill')

# View the cleaned data
stock_data_hist

# Define training and testing areas
train_df = stock_data_hist.iloc[:int(len(stock_data_hist) * 0.9) + 1]  # 90%
test_df = stock_data_hist.iloc[int(len(stock_data_hist) * 0.9):]  # 10%

# Import the required libraries for the model
from statsmodels.tsa.ar_model import AutoReg

# Define training model
model = AutoReg(train_df['Close'], 250).fit(cov_type="HC0")

# Predict data for test data
predictions = model.predict(start=test_df.index[0], end=test_df.index[-1], dynamic=True)

# Predict 90 days into the future
forecast = model.predict(start=test_df.index[0], end=test_df.index[-1] + pd.Timedelta(days=90), dynamic=True)

# View the forecasted values
forecast

# Import matplotlib for plotting
import matplotlib.pyplot as plt

# Seaborn extends the capabilities of Matplotlib
import seaborn as sns
sns.set_style("darkgrid")

# Set fig and ax
fig, ax = plt.subplots()

# Plot the data
ax = train_df.tail(5).plot(ax=ax, color='blue')
ax = test_df.plot(ax=ax, color='orange')
ax = forecast.plot(ax=ax, color='green')
ax = predictions.plot(ax=ax, color='red')

# Customize the plot
plt.legend(["Train Data", "Test Data", "Forecast", "Predictions"])
plt.title("Stock Price Data")

# Show the plot
plt.show()
```

```python
# Import necessary packages
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Load the data from the CSV file
data = pd.read_csv("stock_technical_indicators_last_5_years.csv")
data.dropna(inplace=True)

# Ensure the 'date' column is in datetime format
data['date'] = pd.to_datetime(data['date'])

# Initialize a dictionary to store predictions for each ticker
all_predictions = {}

# List of tickers to process
tickers = data['ticker'].unique()

# Loop through each ticker
for ticker in tickers:
    # Filter data for the current ticker
    ticker_data = data[data['ticker'] == ticker]

    # Set the 'date' column as the index
    ticker_data.set_index('date', inplace=True)

    # Select relevant columns
    ticker_data = ticker_data[['o', 'c', 'h', 'l', 'v', 'SMA', 'EMA', 'RSI', 'ATR']]

    # Prepare data for LSTM
    X = ticker_data[['o', 'h', 'l', 'v', 'SMA', 'EMA', 'RSI', 'ATR']].values
    y = ticker_data['c'].values

    # Normalize the data
    scaler_X = MinMaxScaler(feature_range=(0, 1))
    X_scaled = scaler_X.fit_transform(X)

    scaler_y = MinMaxScaler(feature_range=(0, 1))
    y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))

    # Split the data into training and testing sets
    train_size = int(len(X) * 0.8)
    X_train, X_test = X_scaled[:train_size], X_scaled[train_size:]
    y_train, y_test = y_scaled[:train_size], y_scaled[train_size:]

    # Reshape input to be [samples, time steps, features]
    X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
    X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

    # Define the LSTM model
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=(1, X_train.shape[2])))
    model.add(LSTM(50, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))

    # Compile the model
    model.compile(optimizer='adam', loss='mean_squared_error')

    # Train the model
    model.fit(X_train, y_train, batch_size=1, epochs=10)

    # Predict the stock prices
    predictions = model.predict(X_test)
    predictions = scaler_y.inverse_transform(predictions)

    # Store the predictions in the dictionary
    all_predictions[ticker] = {
        'actual': y[train_size:],  # Actual values
        'predictions': predictions.flatten(),  # Predicted values
    }

    # Plot the results
    train = ticker_data[:train_size]
    valid = ticker_data[train_size:]
    valid['Predictions'] = predictions

    plt.figure(figsize=(16, 8))
    plt.title(f'LSTM Model for {ticker}')
    plt.xlabel('Date')
    plt.ylabel('Close Price USD ($)')
    plt.plot(train['c'])
    plt.plot(valid[['c', 'Predictions']])
    plt.legend(['Train', 'Val', 'Predictions'], loc='lower right')
    plt.show()

# all_predictions now contains the predictions for each ticker
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

# Load the data from the CSV file
data = pd.read_csv("stock_technical_indicators_last_5_years.csv")
data.dropna(inplace=True)

# Ensure the 'date' column is in datetime format
data['date'] = pd.to_datetime(data['date'])

# Set the 'date' column as the index
data.set_index('date', inplace=True)

# Select relevant columns
data = data[['ticker', 'EMA', 'RSI', 'c']]

# Encode the ticker symbols to numerical values
tickers = data['ticker'].unique()
ticker_dict = {ticker: idx for idx, ticker in enumerate(tickers)}
data['ticker'] = data['ticker'].map(ticker_dict)

# Prepare data for LSTM
X = data[['EMA', 'RSI']].values
y = data['c'].values

# Normalize the data
scaler_X = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler_X.fit_transform(X)

scaler_y = MinMaxScaler(feature_range=(0, 1))
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))

# Split the data into training and testing sets
train_size = int(len(X) * 0.8)
X_train, X_test = X_scaled[:train_size], X_scaled[train_size:]
y_train, y_test = y_scaled[:train_size], y_scaled[train_size:]

# Reshape input to be [samples, time steps, features]
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

# Define the LSTM model
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(1, X_train.shape[2])))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, batch_size=1, epochs=5)

# Save the model in Keras format
model.save('lstm_combined_model.keras')

# Load the model (if running separately, use this line)
# model = load_model('lstm_combined_model.keras')

def predict_for_ticker(ticker, data, model, scaler_X, scaler_y, forecast_days=10):
    ticker_idx = ticker_dict[ticker]
    ticker_data = data[data['ticker'] == ticker_idx]
    X_ticker = ticker_data[['EMA', 'RSI']].values
    y_ticker = ticker_data['c'].values

    X_ticker_scaled = scaler_X.transform(X_ticker)
    y_ticker_scaled = scaler_y.transform(y_ticker.reshape(-1, 1))

    X_ticker_scaled = np.reshape(X_ticker_scaled, (X_ticker_scaled.shape[0], 1, X_ticker_scaled.shape[1]))

    predictions_ticker = model.predict(X_ticker_scaled)
    predictions_ticker = scaler_y.inverse_transform(predictions_ticker)

    # Predict future values
    future_predictions = []
    last_sequence = X_ticker_scaled[-1]

    for _ in range(forecast_days):
        next_prediction = model.predict(np.array([last_sequence]))
        future_predictions.append(next_prediction[0])
        # Drop the oldest entry in last_sequence and add the new prediction
        last_sequence = np.append(last_sequence[:, 1:], next_prediction, axis=1)

    future_predictions = scaler_y.inverse_transform(future_predictions)

    return predictions_ticker, y_ticker, future_predictions
```

### FORECASTING

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def forecast_and_plot(ticker, forecast_days, data, model, scaler_X, scaler_y):
    # Predict data for the specified ticker
    predictions, actual, forecast = predict_for_ticker(ticker, data, model, scaler_X, scaler_y, forecast_days)

    # Filter the data for the specific ticker
    ticker_data = data[data['ticker'] == ticker_dict[ticker]]

    if ticker_data.empty:
        print(f"No data available for ticker: {ticker}")
        return

    # Convert predictions and actual values to DataFrames for better handling
    predictions = pd.Series(predictions.flatten(), index=ticker_data.index[:len(predictions)])
    actual = pd.Series(actual.flatten(), index=ticker_data.index[:len(actual)])

    # Create a date range for forecasting, skipping weekends
    forecast_index = []
    last_date = actual.index[-1]
    while len(forecast_index) < forecast_days:
        last_date += pd.Timedelta(days=1)
        if last_date.weekday() < 5:  # Only add weekdays (Monday=0, Sunday=6)
            forecast_index.append(last_date)

    # Convert forecast values to Series
    forecast = pd.Series(forecast.flatten(), index=forecast_index)

    # Plot the results
    plt.figure(figsize=(16, 8))
    plt.title(f'LSTM Model Predictions for {ticker}')
    plt.xlabel('Date')
    plt.ylabel('Close Price USD ($)')
    plt.plot(actual.index, actual, label='Actual', color='blue')
    plt.plot(predictions.index, predictions, label='Predictions', color='orange')
    plt.plot(forecast.index, forecast, label='Forecast', color='green')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.legend(loc='lower right')
    plt.grid()
    plt.tight_layout()
    plt.show()

    # Print the forecasted values for the specified number of days
    print(f"Forecasted values for the next {forecast_days} trading days for {ticker}:")
    for date in forecast_index:
        if date in forecast.index:
            print(f"{date.date()}: {forecast[date]:.2f}")
        else:
            print(f"{date.date()}: Market closed")

# Example usage
ticker = 'MSFT'  # Change this to the desired ticker
forecast_days = 10  # Change this to the desired number of forecast days
forecast_and_plot(ticker, forecast_days, data, model, scaler_X, scaler_y)
```

### Evaluation metrics

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

def evaluate_model(actual, predictions):
    # Calculate Mean Squared Error
    mse = mean_squared_error(actual, predictions)
    
    # Calculate Mean Absolute Error
    mae = mean_absolute_error(actual, predictions)
    
    # Calculate R-squared (R²)
    r_squared = 1 - (mse / np.var(actual))

    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"R-squared (R²): {r_squared:.4f}")

# Example usage for evaluating the model
ticker = 'AAPL'  # Change to desired ticker
predictions_AAPL, actual_AAPL, _ = predict_for_ticker(ticker, data, model, scaler_X, scaler_y, forecast_days=10)

# Flatten the predictions and actual values for evaluation
predictions_AAPL = predictions_AAPL.flatten()
actual_AAPL = actual_AAPL.flatten()

# Evaluate the model's predictions
evaluate_model(actual_AAPL, predictions_AAPL)
```
